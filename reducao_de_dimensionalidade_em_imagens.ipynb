```python
from PIL import Image
import os
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Solicita o caminho da imagem ao usuário
#path = input("Por favor, insira o caminho da imagem: ")
path = './lena.png'

# Verifica se o arquivo existe
if os.path.isfile(path):
    # Abre a imagem
    img = Image.open(path)
    print("Imagem recebida")
else:
    print("Arquivo não encontrado. Por favor, tente novamente.")
```

```python
def converter_para_tons_de_cinza(imagem_colorida):
    largura, altura = imagem_colorida.size
    imagem_tons_de_cinza = Image.new("L", (largura, altura))

    for y in range(altura):
        for x in range(largura):
            pixel = imagem_colorida.getpixel((x, y))
            # Média ponderada dos canais de cor (R, G, B)
            valor_tons_de_cinza = int(0.299 * pixel[0] + 0.587 * pixel[1] + 0.114 * pixel[2])
            imagem_tons_de_cinza.putpixel((x, y), valor_tons_de_cinza)

    return imagem_tons_de_cinza
```

```python
# Converta para tons de cinza
imagem_tons_de_cinza = converter_para_tons_de_cinza(img)

# Exiba a imagem em tons de cinza
display(imagem_tons_de_cinza)
```

```python
def converter_para_imagem_binaria(imagem):
  largura, altura = imagem.size
  imagem_binaria = Image.new("1", (largura, altura))

  for y in range(altura):
    for x in range(largura):
      pixel = imagem.getpixel((x, y))

      # Se estiver em modo "L", apenas um valor de pixel
      valor_pixel = pixel if isinstance(pixel, int) else int((pixel[0] + pixel[1] + pixel[2]) / 3)

      if valor_pixel > 127:
        imagem_binaria.putpixel((x, y), 1)
      else:
        imagem_binaria.putpixel((x, y), 0)

  return imagem_binaria
```

```python
# Converta para tons de cinza
imagem_binaria = converter_para_imagem_binaria(imagem_tons_de_cinza)

# Exiba a imagem em tons de cinza
display(imagem_binaria)
```

```python
# Converta para tons de cinza
imagem_binaria = converter_para_imagem_binaria(img)

# Exiba a imagem em tons de cinza
display(imagem_binaria)
```
